```python
import networkx as nx
import graphviz as gr

G_ = nx.Graph()
```

```python
with open('list.txt') as file:
  text = file.read().split("\n")
for s in text:
  G_.add_node(s)
```

```python
with open('land.txt') as file:
  text = file.read().split("\n")
for s in text:
  v = s.split(" ")
  G_.add_edge(v[0], v[1])


G = G_
G.remove_nodes_from(["UK", "Ireland", "Iceland", "Malta", "Cyprus"])


print("|V| = ", G.number_of_nodes())
print("|E| = ", G.number_of_edges())
```

```python
deq = dict(G.degree)
deq_vertex = list(deq.keys())
deq_deq = list(deq.values())
for i in range(42):
  print("deq(",deq_vertex[i],") =", deq_deq[i])
print("Œî(G) =", max(dict(nx.degree(G)).values()))
print("ùõø(G) =", min(dict(nx.degree(G)).values()))
print(nx.eccentricity(G))
print("rad(G) =", nx.algorithms.radius(G))
print("diam(G) =", nx.algorithms.distance_measures.diameter(G))
print("center(G) =", nx.center(G))
print("all minimum cycles =", nx.minimum_cycle_basis(G))
print("girth(G) =", 3)
print(nx.minimum_edge_cut(G))
print(nx.minimum_node_cut(G))
print("ùúÖ(G) = 1")
print("ùúÜ(G) = 1")
```

```python
import matplotlib.pyplot as plt
import numpy as np

ans = 0

all_cliques = list(nx.find_cliques(G))
maximal_clique = []
for i in all_cliques:
  if(ans < len(i)):
    ans = len(i)
    maximal_clique = i
print(maximal_clique)
```

```python
max_size_of_independent_set = 0
maximal_independet_set = []
for i in range(1000):
    current_independent_set = nx.maximal_independent_set(G)
    if(max_size_of_independent_set < len(current_independent_set)):
       max_size_of_independent_set = len(current_independent_set)
       maximal_independet_set = current_independent_set  
print("|Stable set| =", max_size_of_independent_set)
print("Stable set =", maximal_independet_set)
```

```python
max_size_of_matching = 0
maximal_matching = []

for i in range(1000):
  current_matching = nx.maximal_matching(G)
  if(max_size_of_matching < len(current_matching)):
    max_size_of_matching = len(current_matching)
    maximal_matching = current_matching

print("|Matching| =", max_size_of_matching)
print("Matching =", maximal_matching)
```

```python
min_edge_cover = nx.min_edge_cover(G)
print("|min_edge_cover| =", len(min_edge_cover))
print(min_edge_cover)
```

```python
blocks = nx.biconnected_component_edges(G)

for i in list(blocks):
  print(i)

print(list(nx.articulation_points(G)))
```

```python
components = nx.biconnected_components(G)

for i in list(components):
  print(i)
  
```

```python
Graph_weighted = nx.Graph()

with open('distance.txt') as file:
    text = file.read().split("\n")
    for s in text:
      edge = s.split(" ")
      Graph_weighted.add_edge(edge[0], edge[1], weight = float(edge[2]))

MST = nx.minimum_spanning_tree(Graph_weighted)

plt.figure(figsize=(25,25))
nx.draw_networkx(MST)
```

```python
prufier_code = []
T = nx.Graph()

for edge in MST.edges:
  T.add_edge(list(MST.nodes).index(edge[0]),list(MST.nodes).index(edge[1]))
prufier_code = nx.to_prufer_sequence(T)
print(len(prufier_code))
print(prufier_code)
```
